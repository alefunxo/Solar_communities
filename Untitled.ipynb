```python
def get_bill_hh(g2l,i2g,grid_prices,comm_tariff,export_prices,selection):
    '''
    Description
    -----------
    Calculates the individual bill for every house of the community taking into account the sells to the grid and
    to the community, and the different prices accordingly.
    
    Parameters
    -----------
    g2l (pd.Series): is the grid to load of every single house after their P2P decisions or after self-consume
                    their own PV-generated electricity.
    i2g (pd.Series): is the electricity injected into the grid by every single house after their P2P decisions 
                    or after self-consume their own PV-generated electricity (from the battery or PV).
    
    grid_prices (int or pd.Series): price at which the electricity is paid when comming from the grid 
                                (contrary to comming from the community)
    comm_tariff (pd.Series): price at which the electricity is paid when comming from the community 
                                (at this price is paid the electricity of the exporter and is billed to the imported)
    export_prices (int or pd.Series): price at which the electricity is exported to the grid.
    
    pv_penetration (int): % of pv penetration in the community
    batt_penetration (int): % of battery penetration in the community
    selection (pd.DataFrame): Dataframe including PV sizes, names and selection of the houses
                                with and w/o PV and/or battery according to the pv and battery
                                penetration
    change (): Not used.
    Returns
    -----------
    Total annual cost (float): Total annual cost for every house (including revenue due to sells to the community and/or grid)
    Comments
    -----------
    The electricity produced and consumed inside the community is paid at the community tariff (for the consumer and the producer)
    The electricity comming from the grid is paid at the grid_prices.
    The electricity exported from the community to the grid is paid at the export price]
    TODO
    -----------
    The revenue should be redistributed equally but only among people with PV+Battery??
    double check the prices
    double check the revenue and cost->grid_import*import_price and grid_export*export_price and calculate the rest
    
    
    '''
    #First, calculate the electricity produced by the community that is
    #consumed inside the community (community SC after local SC)
    #This will be paid at the community tariff to the exporter and by
    #the importer
    comm_sc=pd.concat([g2l.sum(axis=1),i2g.sum(axis=1)], axis=1).min(axis=1)
    
    #Second, recalculate the required electricity from the grid
    #This will be paid at the grid tariff
    grid_import=g2l.sum(axis=1)-comm_sc
    #Finally, calculate the export to the grid
    #This will be paid at the export_price
    grid_export=i2g.sum(axis=1)-comm_sc

    #percentage of sc ans ss
    sc_pct=(comm_sc/i2g.sum(axis=1)).fillna(0)
    ss_pct=(comm_sc/g2l.sum(axis=1)).fillna(0)
    
    revenue_comm_sc=pd.DataFrame()
    cost_comm_sc=pd.DataFrame()
    cost_grid=pd.DataFrame()
    revenue_export=pd.DataFrame()
    for i in selection.index:
        #The revenue should be distributed equally among the exporters
        #The cost should be distributed equally among the importer
        revenue_comm_sc[i]=i2g.iloc[:,i]*comm_tariff*sc_pct
        cost_comm_sc[i]=g2l.iloc[:,i]*comm_tariff*ss_pct
        cost_grid[i]=g2l.iloc[:,i]*grid_prices*(1-ss_pct)
        revenue_export[i]=i2g.iloc[:,i]*export_prices*(1-sc_pct)

        
    #return (revenue_comm_sc+revenue_export).sum()
    return (cost_comm_sc-revenue_comm_sc+cost_grid-revenue_export).sum()
    #return [grid_import,grid_export,cost_comm_sc,revenue_comm_sc,cost_grid,revenue_export]
    
```

```python
def plot_graph(df,col,pv=None,batt=None,c2=None,as_today=None):
    '''
    Description
    -----------
    Plots a bar graph of df using the column "col" and grouping by pv and batt.
    If c2 or as_today are not None, they are included in the graph as well.
    Parameters
    -----------
    df (pd.DataFrame): main dataframe to be plotted (usually p2p cluster 1)
    col (str): name of the column to be plotted
    pv (int): % of pv penetration in the community
    batt (int): % of batt penetration in the community
    c2 (pd.DataFrame): dataframe of p2p cluster 2
    as_today (pd.DataFrame): dataframe of today's community
    Returns
    -----------
    None
    
    '''
    
    size=16
    size2=12
    color=['blue', 'cyan','lime', 'red', 'green','orange']
    names=['P2P cluster2','Community\nas today','Community\nP2P cluster 1','Community\nP2P cluster 2']
    df=df.groupby(['pv_penetration',
                      'batt_penetration','use_cluster1_probabilities','kW_dis']).mean().reset_index()
   
    if c2 is not None:
        c2=c2.groupby(['pv_penetration',
                          'batt_penetration','use_cluster1_probabilities','kW_dis']).mean().reset_index()
    if as_today is not None:
        as_today=as_today.groupby(['pv_penetration',
                          'batt_penetration','use_cluster1_probabilities','kW_dis']).mean().reset_index()
    
    ind = np.arange(1,df.shape[0]+1)  # the x locations for the groups
    width = 0.3       # the width of the bars
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    
    
    if pv and batt:
        rects1=ax.bar(ind+width,df[(df.pv_penetration==pv)&(df.batt_penetration==batt)][col].round(0),width,color=color[0],label='P2P cluster1')
        if c2 is not None:
            rects3=ax.bar(ind,c2[(c2.pv_penetration==pv)&(c2.batt_penetration==batt)][col].round(0),width,color=color[2],label='c2')
        if as_today is not None:
            rects2=ax.bar(ind-width,as_today[(as_today.pv_penetration==pv)&(as_today.batt_penetration==batt)][col].round(0),width,color=color[1],label='As today')
        df.loc[(df.pv_penetration==pv)&(df.batt_penetration==batt),['pv_penetration','batt_penetration']]
        tuples=[tuple(x) for x in df.loc[(df.pv_penetration==pv)&(df.batt_penetration==batt),['pv_penetration','batt_penetration']].values]
    elif pv:
        rects1=ax.bar(ind+width,df[(df.pv_penetration==pv)][col].round(0),width,color=color[0],label='P2P cluster1')
        if c2 is not None:
            rects3=ax.bar(ind,c2[(c2.pv_penetration==pv)&(c2.batt_penetration==batt)][col].round(0),width,color=color[2],label='c2')
        if as_today is not None:
            rects2=ax.bar(ind-width,as_today[(as_today.pv_penetration==pv)][col].round(0),width,color=color[1],label='As today')
        df.loc[(df.pv_penetration==pv),['pv_penetration','batt_penetration']]
        tuples=[tuple(x) for x in df.loc[(df.pv_penetration==pv),['pv_penetration','batt_penetration']].values]
    elif batt: 
        rects1=ax.bar(ind+width,df[(df.pv_penetration==pv)&(df.batt_penetration==batt)][col].round(0),width,color=color[0],label='P2P cluster1')
        if c2 is not None:
            rects3=ax.bar(ind,c2[(c2.batt_penetration==batt)][col].round(0),width,color=color[2],label='P2P cluster2')
        if as_today is not None:
            rects2=ax.bar(ind-width,as_today[(as_today.batt_penetration==batt)][col].round(0),width,color=color[1],label='As today')
        df.loc[(df.batt_penetration==batt),['pv_penetration','batt_penetration']]
        tuples=[tuple(x) for x in df.loc[(df.batt_penetration==batt),['pv_penetration','batt_penetration']].values]

    else: 
        rects1=ax.bar(ind+width,df[col].round(0),width,color=color[0],label='P2P cluster1')
        if c2 is not None:
            rects3=ax.bar(ind,c2[col].round(0),width,color=color[2],label='P2P cluster2')
        if as_today is not None:
            rects2=ax.bar(ind-width,as_today[col].round(0),width,color=color[1],label='As today')
        tuples=[tuple(x) for x in df.loc[:,['pv_penetration','batt_penetration']].values]
    plt.draw()
    ax.set_ylabel(col,size=size)
    ax.set_yticklabels([item.get_text() for item in ax.get_yticklabels()],fontsize=size2)
    ax.set_xticks(ind)

    ax.set_xticklabels(tuples,fontsize=size2)
    #labels = [item.get_text() for item in ax.get_xticklabels()]
    ax.set_xlabel('PV and battery \npenetration combination')
    ax.set_title('')
    plt.xticks( rotation='vertical')
    plt.suptitle('')
    if col=='bill':
        ax.axhline(y=0, color='k')
    ax.legend()
```

```python
def plot_boxplot(df,col,pv=None,batt=None):
    '''
    Description
    -----------
    Boxplots of column 'col'. The Dataframe is grouped by pv penetration (pv) and battery penetration (batt) if not None.
    Parameters
    -----------
    df (pd.Series): Single row of the pd.Dataframe (for a single PV and Battery penetration)
    Returns
    -----------
    None
    '''
    size=16
    size2=12
    color=['blue', 'cyan','lime', 'red', 'green','orange']
    names=['Detached','Community\nas today','Community\nP2P cluster 1','Community\nP2P cluster 2']
    if pv and batt:
        bp=df[(df.pv_penetration==pv)&(df.batt_penetration==batt)].round(0).boxplot(by=['pv_penetration',
                      'batt_penetration','use_cluster1_probabilities','kW_dis'], 
                       column=col, 
                       grid=False)
        tuples=[tuple(x) for x in df.loc[(df.pv_penetration==pv)&(df.batt_penetration==batt),['use_cluster1_probabilities','pv_penetration','batt_penetration']].values]
    elif pv:
        bp=df[(df.pv_penetration==pv)].round(0).boxplot(by=['pv_penetration',
                      'batt_penetration','use_cluster1_probabilities','kW_dis'], 
                       column=col, 
                       grid=False)
        tuples=[tuple(x) for x in df.loc[(df.pv_penetration==pv),['use_cluster1_probabilities','pv_penetration','batt_penetration']].values]
    elif batt: 
        bp=df[(df.batt_penetration==batt)].round(0).boxplot(by=['pv_penetration',
                      'batt_penetration','use_cluster1_probabilities','kW_dis'], 
                       column=col, 
                       grid=False)
        tuples=[tuple(x) for x in df.loc[(df.batt_penetration==batt),['use_cluster1_probabilities','pv_penetration','batt_penetration']].values]
    else:
        bp=df.round(0).boxplot(by=['pv_penetration','batt_penetration','use_cluster1_probabilities','kW_dis'], 
                       column=col, 
                       grid=False)
        tuples=[tuple(x) for x in df.loc[:,['use_cluster1_probabilities','pv_penetration','batt_penetration']].values]
    
    plt.draw()
    bp.set_ylabel(col,size=size)
    bp.set_yticklabels([item.get_text() for item in bp.get_yticklabels()],fontsize=size2)
    bp.set_xlabel('cluster, PV and battery \n combination')
    bp.set_xticklabels(tuples,fontsize=size2)

    bp.set_title('')
    plt.xticks( rotation='vertical')
    plt.suptitle('')
    if col=='bill' and not(pv and batt):
        bp.axhline(y=0, color='k')
```

```python
def get_pct_change(current, previous):
    diff = current - previous
    change = 0
    try:
        if diff > 0:
            change = (diff / abs(previous)) * 100
        elif diff < 0:
            diff = previous - current
            change = -((diff / abs(previous)) * 100)
    except ZeroDivisionError:
        return float('inf')
    return change

def get_change(current, previous):
    diff = current - previous
    return diff
```

```python
def get_bill_hh2(g2l,i2g):
    '''
    Description
    -----------
    Calculates the community self-consumption, grid import and export.
    Parameters
    -----------
    g2l (pd.Series): is the grid to load of every single house after their P2P decisions or after self-consume
                    their own PV-generated electricity.
    i2g (pd.Series): is the electricity injected into the grid by every single house after their P2P decisions 
                    or after self-consume their own PV-generated electricity (from the battery or PV).    
    Returns
    -----------
    comm_sc (pd.Series): community self-consumption
    grid_import (pd.Series): Electricity imported from the grid
    grid_export (pd.Series): Electricity exported to the grid
    '''
    comm_sc=pd.concat([g2l.sum(axis=1),i2g.sum(axis=1)], axis=1).min(axis=1)

    #Second, recalculate the required electricity from the grid
    #This will be paid at the grid tariff
    grid_import=g2l.sum(axis=1)-comm_sc
    #Finally, calculate the export to the grid
    #This will be paid at the export_price
    grid_export=i2g.sum(axis=1)-comm_sc
    return [comm_sc,grid_import,grid_export]
```

# Boxplots of individual bills

```python
def boxplot_all(pv,batt,bill_as_today,bill_p2p_c1,bill_p2p_c2,selection):
    ix=bill_detached.loc[(bill_detached.pv_penetration==pv)&(bill_detached.batt_penetration==batt)].index
    df=pd.concat([bill_as_today.loc[ix,:],bill_p2p_c1.loc[ix,:],bill_p2p_c2.loc[ix,:]],ignore_index=True,axis=0).T
    df.columns=['today','c1','c2']
    batt_ix=selection.loc[:,'sub_'+str(pv)+'_'+str(batt)]
    pv_ix=selection.loc[:,'sub_'+str(pv)+'_100']&~batt_ix
    wo_ix=~selection.loc[:,'sub_'+str(pv)+'_100']
    
    if (pv==100)&(batt==100):
        fig, axs = plt.subplots(1,1,figsize=(10, 10),sharey=True)
        l_ix=[batt_ix]
    elif (pv==100):
        fig, axs = plt.subplots(1,2,figsize=(10, 10),sharey=True)
        l_ix=[pv_ix,batt_ix]
    elif (batt==100):
        fig, axs = plt.subplots(1,2,figsize=(10, 10),sharey=True)
        l_ix=[wo_ix,batt_ix]
    else:
        fig, axs = plt.subplots(1,3,figsize=(10, 10),sharey=True)
        l_ix=[wo_ix,pv_ix,batt_ix]
    
    i=0
    for selector in l_ix: 
        if (pv==100)&(batt==100):
            a=df.loc[selector].plot.box(ax=axs)
            axs.set_title('Households with PV and battery')
        elif (pv==100):
            a=df.loc[selector].plot.box(ax=axs[i])
            axs[0].set_title('Households with PV only')
            axs[1].set_title('Households with PV and battery')
        elif (batt==100):
            a=df.loc[selector].plot.box(ax=axs[i])
            axs[0].set_title('Households without PV nor battery')
            axs[1].set_title('Households with PV and battery')
        else:
            a=df.loc[selector].plot.box(ax=axs[i])
            axs[0].set_title('Households without PV nor battery')
            axs[1].set_title('Households with PV only')
            axs[2].set_title('Households with PV and battery')
        a.axhline(y=0, color='k',linestyle='--',alpha=0.5)
        i+=1
    plt.suptitle('Bill per household')
```
